```python
# import packages
import pandas as pd

# read app data as apps
apps = pd.read_csv('datasets/apps.csv')

# read user_reviews.csv as review = pd.read_csv()
review = pd.read_csv('datasets/user_reviews.csv')

# check both df
print('app df: \n', apps.head())
print('review df: \n', review.head())
```

```python
# clean Installs: remove ',' and '+', and convert to integer data type
simbol = [',', '+']

# a for loop to clean
for simbol in simbol:
    apps['Installs'] = apps['Installs'].apply(lambda x: x.replace(simbol, ''))
```

```python
# Find the number of apps in each category, the average price, and the average rating
# select only Category, Price, and Rating columns
app_needed = apps[['Category', 'Price', 'Rating']]

# summarise the info
app_category_info = app_needed.groupby('Category').agg({'Category':'count', 
                                                        'Price':'mean', 
                                                        'Rating': 'mean'})

# change to meaningful column name 
app_category_info.columns = ['Number of apps', 'Average price', 'Average rating']

# free Category as a column from index
app_category_info = app_category_info.reset_index()

app_category_info
```

```python
# Find the top 10 free FINANCE apps having the highest average sentiment score
# merge apps df and review df to combine app category as sentiment score
merge_df = apps.merge(review, on = 'App')

#filter for finance app only
finance_app = merge_df[merge_df['Category'] == 'FINANCE']

# extract needed columns: App and Sentiment Score
score_df = finance_app[['App', 'Sentiment Score']]

# caculate average sentiment score and sort by average sentiment score
user_feedback = score_df.groupby('App').mean().sort_values(by = 'Sentiment Score', ascending = False)

# select only top 10
top_10_user_feedback = user_feedback.head(10).reset_index()
top_10_user_feedback
```
